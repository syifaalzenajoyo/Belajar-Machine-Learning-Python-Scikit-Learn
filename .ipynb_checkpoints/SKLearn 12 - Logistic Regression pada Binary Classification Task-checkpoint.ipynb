# SKLearn 12 - Logistic Regression pada Binary Classification Task

## Formula Dasar
### Simple Linear Regression
- $y = \alpha + \beta x$
- $g(x) = \alpha + \beta x$


### Multiple Linear Regression
- $y = \alpha + \beta_1 x_1 + \beta_2 x_2 + ... + \beta_n x_n$
- $g(X) = \alpha + \beta X$

**Notes:** 
- $x$ pada simple linear regression merepresentasikan feature tunggal
- $X$ pada multiple linear regression merepresentasikan sekumpulan feature ($x_1, x_2$, dst)

### Logistic Regression
- $g(X) = sigmoid(\alpha + \beta X)$
- $sigmoid(x) = \frac{1}{1+exp(-x)}$



**Notes:**
- Rentang nilai pada sumbu x pada ilustarsi di atas hanya sekadar contoh, pada kenyataannya rentang nilai pada sumbu x akan disesuaikan dengan kasus yang dihadapi.
- Rentang nilai pada sumbu y akan selalu berada pada rentang nilai 0 dan 1, rentang nilai tersebut akan berkorelasi dengan nilai binomial probability.
- Logistic regression akan melakukan prediksi nilai yang berada dalam rentang 0 dan 1.
- Pada binary classification task, hanya terdapat 2 buah class. Kedua kelas tersebut dapat kita representasikan dengan 0 dan 1, sehingga jika nilai probabilitynya mendekati 0 akan diklasifikasikan sebagai class 0 dan jika mendekati 1 akan diklasifikasikan sebagai class 1.

## Dataset: SMS Spam Collection Data Set
Source : https://archive.ics.uci.edu/ml/datasets/SMS+Spam+Collection

- Kita akan memanfaatkan logistic regression untuk melakukan prediksi apakah suatu teks sms termasuk ke dalam kategori ham atau spam berdasarkan sms spam collection data set.

```python
import pandas as pd
df = pd.read_csv('C:/Users/ASUS/Desktop/Belajar-Machine-Learning-Python-Scikit-Learn/smsspamcollection/SMSSpamCollection',
                 sep='\t',
                 header=None,
                 names=['label', 'sms'])
df.head()
```

- Proporsi data ham dan spam dalam dataset

```python
df['label'].value_counts()
```

## Training & Testing Dataset

```python
from sklearn.preprocessing import LabelBinarizer

X = df['sms'].values
y = df['label'].values

lb = LabelBinarizer()
y = lb.fit_transform(y).ravel()
lb.classes_
```

**Notes:**
- method ravel() berfungsi untuk mengkonversi multi dimensional array menjadi array 1 dimensi.
- method train_test_split() berfungsi untuk membagi dataset menjadi 2 bagian.
- pemanggilan random_state akan menghasilkan 4 kumpulan nilai yang akan ditampung oleh X_train, X_test, y_train, dan y_test.
- Class 0 merepresentasikan data ham dan class 1 merepresentasikan data spam.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.25,
                                                    random_state=0)

print(X_train, '\n')
print(y_train)
```

## Feature Extraction dengan TF-IDF

- Karena dataset yang kita miliki merupakan dataset teks, kita akan menerapkan TF-IDF.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english')

X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

print(X_train_tfidf)
```

## Binary Classification dengan Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train_tfidf, y_train)
y_pred = model.predict(X_test_tfidf)

for pred, sms in zip(y_pred[:5], X_test[:5]):
    print(f'PRED: {pred} - SMS: {sms}\n')
```

## Evaluation Metrics pada Binary Classification
Berikut ini adalah beberapa evaluation metrics yang dapat kita terapkan untuk mengevaluasi model dari kasus klasifikasi (termasuk binary classification):
- Confusion Matrix
- Accuracy
- Precission & Recall
- F1 Score
- ROC

### Terminologi Dasar
- **True Positive (TP)** : 
    - Sesuatu yang bernilai positif diprediksi secara tepat sebagai positif oleh model. 
    - Pada kasus ini, berarti model telah secara tepat memprediksi data ham sebagai ham dan data spam sebagai spam.
- **True Negative (TN)** : 
    - Sesuatu yang bernilai negatif diprediksi secara tepat sebagai negatif oleh model. 
    - Pada kasus ini, berarti model telah dengan tepat memprediksi data ham sebagai bukan spam dan data spam sebagai bukan ham.
- **False Positive (FP)** : 
    - Sesuatu yang bernilai negatif diprediksi secara tidak tepat sebagai positif oleh model. 
    - Pada kasus ini, berarti model secara keliru telah memprediksi data ham sebagai spam dan data spam sebagai ham
- **False Negative (FN)** : 
    - Sesuatu yang bernilai positif diprediksi secara tidak tepat sebagai negatif oleh model.
    - Pada kasus ini, berarti model secara keliru telah memprediksi data ham sebagai bukan ham dan data spam sebagai bukan spam.

### Confusion Matrix
Confusion matrix juga dikenal sebagai error matrix. Confusion matrix akan menampilkan nilai TP, FP, FN, dan TP.

```python
from sklearn.metrics import confusion_matrix

matrix = confusion_matrix(y_test, y_pred)
matrix
```

```python
tn, fp, fn, tp = matrix.ravel()

print(f'TN: {tn}')
print(f'FP: {fp}')
print(f'FN: {fn}')
print(f'TP: {tp}')
```

```python
import matplotlib.pyplot as plt

plt.matshow(matrix)
plt.colorbar()

plt.title('Confusion Matrix')
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

**Notes:**
- Sumbu x merepresentasikan predicted label
    - Nilai 0 pada sumbu x merepresentasikan nilai diprediksi sebagai class 0
    - Nilai 1 pada sumbu x merepresentasikan nilai diprediksi sebagai class 1
- Sumbu y merepresentasikan true label
    - Nilai 0 pada sumbu y merepresentasiikan bahwa nilai tersebut merupakan bagian dari class 0
    - Nilai 1 pada sumbu y merepresentasiikan bahwa nilai tersebut merupakan bagian dari class 1
- Hubungan sumbu x dan sumbu y (x,y)
    - Nilai (0,0) merepresentasikan bahwa nilai tersebut direpresentasikan sebagai class 0 dan memang sebenarnya merupakan bagian dari class 0. Pada kasus ini 0 merepresentasikan data bukan spam (ham), sehingga (0,0) berarti data bukan spam diprediksi sebagai bukan spam (TN).
    - Nilai (0,1) merepresentasikan bahwa nilai tersebut direpresentasikan sebagai class 0 ,namun sebenarnya merupakan bagian dari class 1. Pada kasus ini 0 merepresentasikan data bukan spam (ham) dan 1 merepresentasikan spam, sehingga (0,1) berarti model secara keliru memprediksi data sebagai bukan spam (ham) padahal sebenarnya spam (FP).
    - Nilai (1,0) merepresentasikan bahwa nilai tersebut direpresentasikan sebagai class 1 ,namun sebenarnya merupakan bagian dari class 0. Pada kasus ini 0 merepresentasikan data bukan spam (ham) dan 1 merepresentasikan spam, sehingga (1,0) berarti model secara keliru memprediksi data sebagai spam padahal sebenarnya bukan spam (ham) (FN).
    - Nilai (1,1) merepresentasikan bahwa nilai tersebut direpresentasikan sebagai class 1 dan memang sebenarnya merupakan bagian dari class 1. Pada kasus ini 1 merepresentasikan spam, sehingga (1,1) berarti data spam diprediksi sebagai spam (TP).

### Accuracy
Accuracy mengukur porsi dari hasil predikasi yang tepat. Accuracy menilai banyaknya data yang diprediksi secara tepat dari banyaknya keseluruhan prediksi. 

$Accuracy = \frac{TP + TN}{TP+TN+FP+FN} = \frac{correct}{total}$

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

### Precission & Recall
Performa suatu classifier juga dapat diukur berdasarkan nilai Precission dan Recall.

**Precission or Positive Value (PPV)**

$Precission = \frac{TP}{TP+FP}$

```python
from sklearn.metrics import precision_score

precision_score(y_test, y_pred)
```

**Recall or True Positive Rate (TPR) or Sensitivity**

$Recall = \frac{TP}{TP+FN}$

```python
from sklearn.metrics import recall_score

recall_score(y_test, y_pred)
```

### F1-Score
F1-score atau F1-measure merupakan harmonic mean dari precission dan recall.

$F1-score = \frac{precission \times recall}{precission + recall}$

```python
from sklearn.metrics import f1_score

f1_score(y_test, y_pred)
```

### ROC: Receiver Operating Characteristic
ROC menawarkan visualisasi terhadap performa dari classifier dengan membandingkan nilai Recal (TPR) dengan nilai Fallout (FPR).

$fallout = \frac{FP}{TN + FP}$



```python
from sklearn.metrics import roc_curve, auc

prob_estimates = model.predict_proba(X_test_tfidf)

fpr, tpr, threshhold = roc_curve(y_test, prob_estimates[:, 1])
nilai_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, 'b', label=f'AUC={nilai_auc}')
plt.plot([0,1], [0,1], 'r--', label='Random CLassifier')

plt.title('ROC: Receiver Operating Characteristic')
plt.xlabel('Fallout or False Positive Rate')
plt.ylabel('Recall or True Positive Rate')
plt.legend()
plt.show()
```
