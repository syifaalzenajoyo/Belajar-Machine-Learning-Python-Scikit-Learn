## Konsep Dasar

### Terminology: Root Node, Internal Node, Leaf Node


- Pada struktur decision binary tree, setiap node terdiri atas 2 cabang pilihan.
- **Root Node**
    - Node yang terletak pada posisi paling atas. 
    - Pada contoh di atas, yang menjadi root node adalah `Fear Technology?`
- **Internal Node**
    - Node yang berada di antara root node dan leaf node
    - Pada contoh di atas, yang menjadi internal node adalah `Your dad rich?`, `Care about privacy?`, dan `Have life?`
- **Leaf Node**
    - Node yang terletak pada posisi ujung paling bawah dan menjadi prediksi class yang dihasilkan dari struktur decision binary tree. 
    - Pada contoh di atas, yang menjadi leaf node adalah `Apple`, `Chrome`, `Ubuntu Linur`, `Arch Linux`, dan `Windows`. 
- Algorithm Decision Tree:
    - CART (Classification and Regression Tree), default yang digunakan pada SKLearn
    - ID3
    - C4.5
    - C5.0

### Gini Impurity


- Gini impurity digunakan untuk melakukan pengukuran ketidakmurnian (Impurity Measure). 
- Gini impurity memiliki jangkauan nilai 0-1.
    - 0 merepresentasikan nilai murni yang sempurna
    - 1 merepresenatasikan nilai paling tidak murni
- Formula dari Gini Impurity:
\begin{align*}
G &= 1 - \sum_i^n P_i^2
\end{align*}

Pada studi kasus ini, terdapat 2 class (hijau dan biru). Kemudian kita melakukan splitting, sehingga terdapat 2 ruas (kiri dan kanan) 
#### Ruas Kiri
- Pada ruas kiri, hanya terdiri dari 1 class saja yaitu class biru. 
- Nilai Gini Impurity Ruas Kiri didapat dari 1 dikurangi oleh hasil kuadrat dari nilai probability class biru.
- Nilai probality class biru pada ruas kiri adalah 4/4 (banyaknya anggota class biru yang ada pada ruas tersebut / keseluruhan anggota class pada ruas tersebut).
\begin{align*} 
G &= 1 - \sum_i^n P_i^2 \\
  &= 1 - P(biru)^2 \\
  &= 1 - (\frac{4}{4})^2 = 0
\end{align*}
- Dihasilkan nilai G = 0, artinya ruas kiri memiliki nilai gini impurity yang murni sempurna.

#### Ruas Kanan:
- Pada ruas kanan, terdapat 2 class yaitu class biru dan hijau.
- Nilai Gini Impurity Ruas Kanan didapat dari 1 dikurangi oleh total nilai hasil kuadrat dari nilai probability class biru dan nilai hasil kuadrat dari nilai probability class hijau.
- Nilai probality class biru pada ruas kanan adalah 1/6 (banyaknya anggota class biru yang ada pada ruas tersebut / keseluruhan anggota class pada ruas tersebut).
- Nilai probality class hijau pada ruas kanan adalah 5/6 (banyaknya anggota class hijau yang ada pada ruas tersebut / keseluruhan anggota class pada ruas tersebut).
\begin{align*}
G &= 1 - \sum_i^n P_i^2 \\
  &= 1 - (P(biru)^2 + P(hijau)^2)\\
  &= 1 - ( (\frac{1}{6})^2 + (\frac{5}{6})^2 ) = 0.278
\end{align*}

#### Average Gini Impurity:
- Nilai rata-rata gini impurity pada kasus kali ini diperoleh dari hasil penjumlahan (banyaknya anggota pada ruas kiri, dibagi dengan banyaknya seluruh anggota, kemudian dikalikan dengan nilai gini impurity dari ruas kiri) dengan (banyaknya anggota pada ruas kanan, dibagi dengan banyaknya seluruh anggota, kemudian dikalikan dengan nilai gini impurity dari ruas kanan).
\begin{align*}
G &= \frac{4}{4+6} \times 0 + \frac{6}{4+6} \times  0.278 \\
  &= 0.1668
\end{align*}

### Information Gain


- Information Gain diperoleh dari hasil pengurangan nilai Gini Impurity sebelum dilakukan splitting oleh Nilai rata-rata dari Gini Impurity setelah dilakukan splitting.

### Membangun Decision Tree



- Terdapat dataset yang terdiri atas 5 baris dan 3 kolom
- Kolom Color dan Diameter akan dijadikan sebagai features
- Kolom Label akan dijadikan sebagai target label
- Untuk membuat decision tree dari kasus di atas yaitu melakukan prediksi label berdasarkan color dan diameternya:
    1. Menghitung nilai gini impurity sebelum dilakukan splitting.
    2. List semua kemungkinan pertanyaan untuk splitting.
    3. Melakukan splitting berdasarkan kemungkinan pertanyaan yang telah dibuat satu per satu.
    4. Menghtiung nilai ginii impurity setelah splitting dan nilai information gain.
    5. Decision tree yang akan dipilih adalah decision tree dengan nilai information gain tertinggi.
- Nilai dari gain impurity sebelum dilakukan splitting.   
\begin{align*} 
G &= 1 - (P(apple)^2 + P(grape)^2 + P(lemon)^2) \\
   &=1 - ( (\frac{2}{5})^2 + (\frac{2}{5})^2 + (\frac{1}{5})^2 ) \\
   &= 0.63
\end{align*}

## Dataset

- Dataset yang akan digunakan adalah Iris Dataset yang berisikan data terkait bunga iris yang terdiri dari 3 spesies, yaitu `Iris Setosa`, `Iris Virginica`, dan `Iris Versicolor`.
- Dataset tersebut akan displit menjadi 2 bagian yaitu nilai feature yang akan ditampung oleh variabel X dan target labelnya yang akan ditampung oleh variabel y.

```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)

print(f'Dimensi Feature: {X.shape}')
print(f'Class: {set(y)}')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=0)
```

## Classification dengan DecisionTreeClassifier
- Parameter `max_depth=4` pada object DecisionTreeClassifier() merepresentasikan bahwa DecisionTreeClasifier() yang dihasilakn memiliki kedalaman maximum sebanayak 4 layer.

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=4)

model.fit(X_train, y_train)
```

## Visualisasi Model

```python
import matplotlib.pyplot as plt
from sklearn import tree

plt.rcParams['figure.dpi'] = 85
plt.subplots(figsize=(10, 10))
tree.plot_tree(model, fontsize=10)
plt.show()
```

## Evaluasi Model

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```
